객체와 클래스

객체의 개념
- 객체 지향 프로그래밍(Object Oriented Programming)
 - 객체(object): 현실 세계에 존재하는 모든 것
 - 속성(attribute): 객체의 상태를 의미, 변수와 유사, 객체의 변수
 - 메서드(method): 객체의 동작을 의미, 함수와 유사, 객체의 함수

- balance: 잔액 저장 변수
- deposit: 계좌에 입금하는 함수
- withdraw: 계좌에서 출금하는 함수
- balance: 함수는 외부에 생성
 - global 키워드로 외부 함수의 변수를 사용

<지역변수 & 전역변수>
- 지역변수: 함수 안에 만들어진 변수, 함수가 호출될 때 사용, 끝날 때 없어짐
- 전역변수: 함수 밖에 만들어진 변수, 함수와 관계없이 동작

```python
balance = 0

def deposit(amount):
 global balance
 balance += amount
 return balance

def withdraw(amount):
 global balance
 balance -= amount
 return balance

print(deposit(100))
print(withdraw(30))
```

클래스의 개념
- 객체는 클래스를 이용해서 생성
- 클래스란 객체를 생성하는 틀(template)
- 인스턴스(lnstance): 클래스로부터 만들어진 객체

파이썬의 클래스와 객체 생성

클래스의 생성
- 은행 계좌 클래스: Account
 - 속성: balance(잔액)
 - 동작: deposit(입금), withdraw(출금)

```python
class Account:               # 클래스 이름의 첫 글자는 대문자
 balance = 0                 # 잔액 속성 balance 초기값 0으로 셋팅

 def deposit(self, amount):  # method의 첫 번째 매개 변수는 self로 주어짐
   self.balance += amount    # 클래스의 balance에 접근시에는 self.balance로 속성 앞에 self.를 붙여야 함
   return self.balance

 def withdraw(self, amount):
   self.balance -= amount
   return self.balance
```

객체의 생성

```python
a1 = Account()
```

- 객체 지향 프로그래밍은 객체의 속성과 메서드를 하나의 캡슐화(encapsulation)이 핵심
- 내부 구현은 숨기고, 외부에는 속성과 메서드에 접근하는 인터페이스만 제공
- 캡술화를 통해 코드의 안전성을 높이고, 유지 보수를 용이하게 함
- 외부에서 속성을 접근하지 못하도록 private 키워드로 설정할 수 없게 함
- 속성 앞에 _ 를 붙이면 그 속성은 private 속성이 됨

```python
class Account:
 def __init__(self, amount=0):
   self.__balance = amount

 def deposit(self, amount):
   self.__balance += amount
   return self.__balance

 def withdraw(self, amount):
   self.__balance -= amount
   return self.__balance

a1 = Account(50)
print(a1.__balance)
```

private 속성 값 얻기
- 속성 값을 반환하는 메서드를 이용
- get_balance

```python
class Account:
 def __init__(self, amount=0):
   self.__balance = amount     # private 속성

 def deposit(self, amount):
   self.__balance += amount
   return self.__balance

 def withdraw(self, amount):
   self.__balance -= amount
   return self.__balance

 def get_balance(self):
   return self.__balance

a1 = Account(50)
print(a1.get_balance())         # get_속성이
print(a1.deposit(100))
```

LAB. 원 클래스와 객체
- _radius: 반지름
- det_radius(): 반지름 확인 메서드
- set_radius(): 반지름 설정 메서드
- get_area(): 원의 넓이를 구하는 메서드, A = 파이알제곱
- get_circumference: 원의 둘레를 구하는 메서드, 원둘레 = 2 * 파이 * 반지름

```python
import math
```

```python
class Circle:
 def __init__(self, radius=10):
   self.__radius = radius

 def get_radius(self):
   return self.__radius

 def set_radius(self, radius):
   self.__radius = radius

 def get_area(self):
   return math.pi*self.__radius**2

 def get_circumference(self):
   return 2*math.pi*self.__radius
```

```python
c1 = Circle(7)
print('넓이:', c1.get_area())
print('둘레:', c1.get_circumference())
```

```python
c1.set_radius(10)
print('넓이:', c1.get_area())
print('둘레:', c1.get_circumference())
```

LAB.학생 클래스와 객체
- Student: 학생 클래스
 - __name: 이름 속성
 - __math: 수학 성적 속성
 - __computer: 컴퓨터 성적 속성
 - get_name(): 이름을 확인 하는 메서드
 - get_average(): 성적 평균을 설정하는 메서드
 - set_math(): 수학 성적을 설정하는 메서드
 - set_computer(): 컴퓨터 성적을 설정하는 메서드

```python
class Student:

 def __init__(self, name, math, computer):
   self.__name = name
   self.__math = math
   self.__computer = computer

 def get_name(self):
   return self.__name

 def get_average(self):
   return (self.__math + self.__computer)/2

 def set_math(self, math):
   self.__math = math

 def set_computer(self, computer):
   self.__computer = computer
```

```python
s1 = Student('hanbit', 95, 89)
print(s1.get_name(), s1.get_average())
```

```python
s1.set_computer(97)
print(s1.get_name(), s1.get_average())
```

상속(inheritance)

상속의 개념
- 기존 클래스의 속성과 메서드를 공뷰하는 메커니즘
- 새로운 속성과 메서드를 추가할 수 있음
- 상속을 통해 정의되는 클래스: 자식 클래스(child class) or subclass
- 상속을 해주는 클래스: 부모 클래스(parent class) or superclass

상속과 생성자

```python
class Student:
 def __init__(self, name, computer):
   self.name = name
   self.computer = computer

 def set_computer(self, computer):
   self.computer = computer
 def get_name(self):
   return self.name
```

```python
class Science(Student):
 def __init__(self, name, computer, science):
   super().__init__(name, computer)            # super(). 부모클래스 생성자 호출
   self.science = science

 def set_science(self, science):
   self.science = science

 def get_average(self):
   return (self.science+self.computer)/2
```

```python
str1 = Science("홍길동", 90, 80)
print(str1.get_name())
print(str1.get_average())
```

```python
class Liberalarts(Student):
 def __init__(self, name, computer, social):
   super().__init__(name, computer)
   self.social = social

 def set_social(self, social):
   self.social = social

 def get_average(self):
   return (self.social+self.computer)/2
```

```python
st2 = Liberalarts('hanbit2', 98, 88)
print(st2.get_average())
st2.set_social(96)
print(st2.get_name(), ':', st2.get_average())
```

LAB. 도형 클래스를 상속받는 원가 직사각형 클래스의 객체 생성
- Figuer 클래스를 상속 받는 Circle class와 Rectangle class 생성
- Figuer class
 - area 속성: 넓이를 의미
 - preimeter(): 둘레를 의미
 - get_area() : 넓이를 반환하는 메서드
get_perimeter() : 둘레를 반환하는 메서드
- Circle class
 - radius : 반지름 속성
 - set_radius() : 반지름을 설정하는 메서드
 - get_radius() : 반지름을 반환하는 메서드
- Rectangle
 - width : 가로를 의미하는 속성
 - height : 세로를 의미하는 속성
 - set_width() : 가로를 설정하는 메서드
 - set_height() : 세로를 설정하는 메서드
 - get_width() : 가로를 반환하는 메서드
 - get_height() : 세로를 반환하는 메서드

```python
class Figure:
 def __init__(self, area, perimeter):
   self.area = area
   self.perimeter = perimeter

 def get_area(self):
   return self.area

 def get_perimeter(self):
    return self.perimeter
```

```python
class Circle(Figure):
 def __init__(self, radius=10):
   area = 3.14*radius**2
   perimeter = 2*3.14*radius
   super().__init__(area, perimeter)
   self.radius = radius

 def set_radius(self, radius):
   self.radius = radius
   self.area = 3.14*radius**2
   self.perimeter = 2*3.14*radius

 def get_radius(self):
   return self.radius
```

```python
class Rectangle(Figure):
 def __init__(self, width=10, height=10):
   area = width*height
   perimeter = 2*width+2*height
   super().__init__(area, perimeter)
   self.width = width
   self.height = height

 def set_width(self, width):
   self.width = width
   self.area = width*self.height
   self.perimeter = 2*width+2*self.height

 def set_height(self, height):
   self.height = height
   self.area = self.width*height
   self.perimeter = 2*self.width+2*height

 def get_width(self):
   return self.width

 def get_height(self):
   return self.height
```

```python
f1 = Circle(7)
print(f1.get_area(), f1.get_perimeter())
f1.set_radius(10)
print(f1.get_area(), f1.get_perimeter())
```

```python
f2 = Rectangle(5, 4)
print(f2.get_area(), f2.get_perimeter())
f2.set_width(7)
print(f2.get_area(), f2.get_perimeter())
```
