응용: dictionary의 key와 values값을 list로 변환하여 그래프 그리기

LAB 혈액형 빈도수 시각화하기
- 혈액형을 입력받고 빈도수를 시각적으로 비교
- 종료가 입력될 때까지 혈액형을 입력받아 딕셔너리로 처리

```python
import matplotlib.pyplot as plt

blood = {'A':0,'B':0,'O':0,'AB':0}

while True:
 s = input('혈액형(A, B, O, AB) 또는 종료: ')
 if s=='종료':
   break
 elif s in blood:
   blood[s] += 1
 else:
   print('잘못 입력했습니다.')

plt.figure(figsize=(5,3))
plt.bar(blood.keys(), blood.values(), width=0.6)
plt.xlabel('blood type')
plt.ylabel('frequency')
plt.show()
```

LAB 난수를 이용하여 히스토그램 그리기

```python
import matplotlib.pyplot as plt
import random

x = [random.normalvariate(0, 1) for _ in range(1000)] # i 변수 대신 기호(_)를 사용하기 도 함.

plt.figure(figsize=(5,3))
plt.ylabel('frequency')
plt.hist(x, bins=20, alpha=0.5)
plt.show()
```

인기강좌의 빈도수 그래프
- 영어학원에서 수강 인원에 따라 강의실 배정하기 위해 수강 희망 강좌를 조사
- 수강을 희망하는 영어 강좌 조사 결과를 그래프로 시각화하는 프로그램

1. 인기 영어 강좌릐 진도수 구하기
- 수강하기 원하는 영어 강좌를 반복적으로 입력 받아 강좌에 대한 빈도수를 구하는 프로그램
- 키는 강좌명, 값은 0인 course 딕셔너리를 생성
- 값은 빈도수

```python
course ={'TOEIC':0, 'TOEFL':0, 'TOEIC SPEAKING':0, 'OPIC':0, 'IELTS':0}

print('수강하고 싶은 강좌를 입력하세요.')
print('복수 개의 강좌를 입력하려면 콤마로 구분하세요.')

while True:
 s = input('TOEIC, TOEFL, TOEIC SPEAKING, OPIC, IELTS, 종료: ')
 if s=='종료':
   break
 for a in s.split(','):
   if a in course:
     course[a] += 1
   else:
     print(f'{a}는 등록되지 않은 강좌입니다.')

for key, value in course.items():
 print(key, value)
```

2. 빈도수 그래프 그리기
- 강좌에 비니도수를 막대 그래프로 그리기
- bar 함수의 첫 번째 매개변수인 항목 이름은 course.keys()로 구하고
- bar 함수의 두 번째 매개변수인 항목 값은 course.values()로 구함

```python
course ={'TOEIC':0, 'TOEFL':0, 'TOEIC SPEAKING':0, 'OPIC':0, 'IELTS':0}
print('수강하고 싶은 강좌를 입력하세요.')
print('복수 개의 강좌를 입력하려면 콤마로 구분하세요.')
while True:
 s = input('TOEIC, TOEFL, TOEIC SPEAKING, OPIC, IELTS, 종료: ')
 if s=='종료':
   break
 for a in s.split(','):
   if a in course:
     course[a] += 1
   else:
     print(f'{a}는 등록되지 않은 강좌입니다.')
plt.figure(figsize=(5.5,3))
plt.bar(course.keys(), course.values(), color='g', width=0.6)
plt.title('preferred english course')
plt.xlabel('english course')
plt.ylabel('frequency')
plt.show()
```
