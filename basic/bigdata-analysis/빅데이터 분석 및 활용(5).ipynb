데이터 읽어 오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

raw_data = pd.read_csv()

raw_data.info()
```

```python
raw_data.head()
```

```python
필요한 열만 추출하기
- Age: 나이
- Country: 국가 정보
- LanguageHaveWorkedWith: 다를 줄 아는 프로그래밍 언어
- LearnCode: 프로그래밍을 배운 방법
```

```python
 # column에 특정 열 이 있는지 검색
print('Country' in raw_data.columns)
print('LanguageHaveWorkedWith' in raw_data.columns)
print('LearnCode' in raw_data.columns)
```

```python
revised_data = raw_data[["Age", "Country", "LanguageHaveWorkedWith", "LearnCode"]]

revised_data.head()
```

많이 사용하는 언어 그래프 그리기
- 프로그래밍 언어 데이터 추출: LanguageHaveWorkedWith

```python
languages = revised_data['LanguageHaveWorkedWith']
languages
```

단일 객체를 리스트로 변환
- split(): 프로그래밍 언어별 그룹화를 위해서 각 프로그래밍 언어를 하나의 값으로 쪼개야 함

```python
languages = languages.str.split(";")
languages
```

리스트의 각 항목을 열로 나누기
- explode(): 리스트의 항목을 행으로 나누는 메서드

```python
exploded_languages = languages.explode()
exploded_languages
# 한 애가 4개의 언어를 말하고 있음, 왜 327925인지 이해해야함
```

프로그래밍별 응답 수 확인

```python
size_by_laguages = exploded_languages.groupby(exploded_languages).size()
size_by_laguages
```

프로그래밍별 응답 수 확인

```python
size_by_laguages.nlargest(10).plot.pie(figsize=(10, 10))
```

백분율 표기

```python
size_by_laguages.nlargest(10).plot.pie(figsize=(10, 10), autopct='%1.1f%%')
```

원하는 연령대 데이터만 추출해 그래프 그리기

```python
 # 25~34세 연령대가 사용한 프로그래밍 언어 정보 추출
languages_for_25_34 = revised_data[revised_data.Age == '25-34 years old']["LanguageHaveWorkedWith"]
 # 데이터 확인
languages_for_25_34
```

```python
 # 상위 10개 데이터 파이 그래프 그리기
languages_for_25_34.groupby(languages_for_25_34).size().nlargest(10).plot.pie(figsize=(10, 10))
```
