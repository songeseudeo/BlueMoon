bottom에 값을 spring + summer 설정하기 위해서는 배열이어야 함

```python
# import numpy as np
# import matplotlib.pyplot as plt

year = [2017, 2018, 2019, 2020, 2021]
spring = np.array([124.9, 383.5, 175.0, 173.7, 330.5])
summer = np.array([612.7, 620.6, 508.2, 1037.6, 612.8])
autumn = np.array([177.9, 351.3, 440.8, 270.4, 256.4])
winter = np.array([75.2, 68.7, 168.8, 47.8, 13.3])

plt.bar(year, spring, label='spring')
plt.bar(year, summer, bottom=spring, label='summer')
plt.bar(year, autumn, bottom=spring+summer, label='autumn')
plt.bar(year, winter, bottom=spring+summer+autumn, label='winter')
plt.title('precipitation')
plt.xlabel('year')
plt.ylabel('mm')
plt.legend()
plt.show()
```

Numpy 수학 함수
- add(): 두 배열에서 같은 위치의 요소끼리 더한다.
- subtract(): 첫 번째 배열의 요소에서 두 번째 배열의 요소를 뺀다.
- multiply(): 두 배열에서 같은 위치의 요소끼리 곱한다.
- divide(): 첫 번째 배열의 요소를 두 번째 배열의 요소로 나눈다.
- floor_divide(): 첫 번째 배열의 요소를 두 번째 배열의 요소로 나눈 몫을 구한다.
- mod(): 첫 번째 배열의 요소를 두 번째 배열의 요소로 나눈 나머지를 구한다.
- power(): 첫 번째 배열의 요소를 두 번째 배열의 요소만큼 제곱한다.
- abs(): 배열 각 요소의 절댓값을 구한다.
- sqrt(): 배열 각 요소의 제곱근을 구한다.
- square(): 배열 각 요소의 제곱을 구한다.
- exp(): 배열 각 요소의 지수 e^x를 구한다.
- log(), log2(), lod10(): 배열 각 요소의 자연로그, 로그2, 로그10을 구한다.
- cos(): 배열 각 요소의 코사인 값을 구한다.
- sin(): 배열 각 요소의 사인 값을 구한다.
- tan(): 배열 각 요소의 탄젠트 값을 구한다

- sin 함수와 cos 함수

```python
x = np.linspace(-np.pi, np.pi*2, 100)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x), label='sin')
plt.legend()
plt.subplot(2,1,2)
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.show()
```

통계함수
- sum(): 배열 각 요소의 합을 구한다.
- mean(): 배열 각 요소의 평균을 구한다.
- std(), var(): 배열 각 요소의 표준편차, 분산을 구한다.
- max(). min(): 배열의 최댓값, 최솟값을 찾는다.
- argmax(), argmin(): 배열 최댓값의 인덱스, 최솟값의 인덱스를 찾는다.
- median(): 배열의 중앙값을 찾는다.
- cumsum(): 배열 각 요소의 누적 합을 구한다.
- cumprod(): 배열 각 요소의 누적 곱을 구한다.

난수 생성 함수
- randint(): 설정한 구간 사이의 정수로 된 난수를 생성한다.
- rand(): 균등분포를 따르는 0~1 사이의 난수를 생성한다.
- randn(): 평균이 0이고 분산이 1인 표준정규분포를 따르는 난수를 생성한다.
- normal(): 설정한 평균과 표준편차인 정규분포를 따르는 난수를 생성한다.
- uniform(): 균등분포를 따르는 설정한 구간 사이의 난수를 생성한다.
- seed(): 시드를 설정한다. 시드가 같으면 동일한 난수가 생성된다.

```python
x = np.random.randn(1000)

plt.figure(figsize=(5, 3))
plt.hist(x, bins=40, alpha=0.5)
plt.ylabel('frequency')
plt.show()
```

where 함수
- 조건이 참인지 거짓인지에 따라 다른 값을 선택하여 반환
- 배열의 모든 요소를 처리할 수 있음
 - np.where(조건식, 값1, 값2)
 - 참이면 값1, 거짓이면 값2

```python
score = np.array([71, 80, 60, 90, 65])
result = np.where(score>=90, 'A',
                 np.where(score>=80, 'B',
                          np.where(score>=70, 'C', 'D')))
print(result)
```

LAB 코로나19 확진자 수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

months = np.arange(1, 13)
num_cases = [11, 2920, 6636, 750, 615, 1141, 1044, 5473, 3707, 2444, 7324, 26198]
cum_cases = np.cumsum(num_cases) # 배열에서 주어진 축에 따라 누적되는 원소들의 누적 합을 계산하는 함수

plt.figure(figsize=(5, 3))
plt.plot(months, num_cases, label='number of cases')
plt.plot(months, cum_cases, label='accumulated number of cases')
plt.xlabel('month')
plt.legend()
plt.show()
```

LAB 동전 던지기 시뮬레이션
- 동전을 던졌을 때 앞면이 나올 확률은 50%

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
for i in range(3):
 coin = np.random.randint(0, 2, 5000)
 ind = np.arange(1, len(coin)+1)   # 1은 앞면, 0은 뒷면 이라 가
 p_c = np.cumsum(coin)/ind*100
 plt.plot(ind, p_c)
plt.ylim(0,100)
plt.ylabel('%')
plt.show()
```

출생아 수 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
year = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
boy = [248958,223883,223356,224906,208064,184308,167686,155416,139362,133516]
girl = [235592,212572,212079,213514,198179,173463,159136,147260,132975,127046]
```

```python
ind = np.arange(len(year))

w = 0.4
plt.figure(figsize=(6, 3.5))
plt.bar(ind-w/2, boy, color='b', width=w, label='boy')
plt.bar(ind+w/2, girl, color='r', width=w, label='girl')
plt.xticks(ind, year)
plt.legend()
plt.xlabel('year')
plt.ylabel('number of births')
plt.show()
```
