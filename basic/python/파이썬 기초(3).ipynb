함수를 이용해서 사칙연산하기

```python
def add(x, y):
  return x+y
def sub(x, y):
  return x-y
def mul(x, y):
  return x*y
def div(x, y):
  return x/y

a = int(input('정수: '))
b = int(input('정수: '))
print(f'{a}+{b}={add(a, b)}')
print(f'{a}-{b}={sub(a, b)}')
print(f'{a}*{b}={mul(a, b)}')
print(f'{a}/{b}={div(a, b): .2f}')
```

함수 하나로 몫과 나머지 반환하기

```python
def div(x, y):
  return x//y, x%y

a = int(input('정수: '))
b = int(input('정수: '))
quotient, remainder = div(a, b)
print(f'{a}를 {b}로 나눈 몫은 {quotient}이고 나머지는 {remainder}')
```
