리스트의 개념

- 리스트(list): 여러 개의 데이터를 저장하는 공간
 - 예) 값 5개를 저장하려면 변수 5개를 만듦
 - 리스트 하나를 만들어 5개의 값을 한 번에 저장할 수 있음

<리스트 개념이 없을 경우>
- 5개의 가격을 저장
- 5개의 가격이 5% 인상되었다면

```python
price1 = 1500
price2 = 3200
price3 = 2500
price4 = 4000
price5 = 5500

price1 = price1 + price1 * 0.05
price2 = price2 + price2 * 0.05
price3 = price3 + price3 * 0.05
price4 = price4 + price4 * 0.05
price5 = price5 + price5 * 0.05

print(price1)
print(price2)
print(price3)
print(price4)
print(price5)
```

<리스트를 사용하는 경우>
- 대괄호([]) 안에 데이터를 콤마 (,)로 구분하여 입력
- 각 데이터를 항목(item) 또는 요소(element)라 함

- 인덱스(index): 대괄호 안에 표시되는 숫자, 시작은 0부터

```python
prices = [1500, 3200, 2500, 4000, 5500]
print(prices)

for i in range(len(prices)):
  prices[i] = prices[i] + prices[i] * 0.05

print(prices)
```

```python
i = 0 #다른 초기값을 주지 않으면 어차피 0임
for i in range(len(prices)): #i가 의미하는 것은 초기값 #range 증가값 함수 #len 범위를 나타내는 함수
  prices[i] = prices[i] + prices[i] * 0.05

print(prices)
```

- [식 for 변수 in 반복 가능한 것 if 조건식]

```python
sequence = []
for x in range(1, 19):
  if x % 7 == 0:
    sequence.append(x)
print(sequence)
```

```python
sequence = [x for x in range(1, 19) if x % 7 == 0]
print(sequence)
```

- 리스트 안 성적이 70 이상이면 합격 아니면 불합격

```python
score = [70, 80, 90, 65, 100]
result = []

for a in score:
  if a >= 70:
    result.append('합격')
  else:
    result.append('불합격')

print(result)
```

```python
score = [71, 80, 60, 90, 65]
result = []

for a in score:
  if a >= 70:
    result.append('합격')
  else:
    result.append('불합격')

print(result)
```

```python
score = [71, 80, 60, 90, 65]
result = ['합격' if a >= 70 else '불합격' for a in score]
print(result)
```

pop()

reverse()
- 리스트 항목의 값은 그대로 두면서 순서만 재배열
- 리스트.reverse()

```python
nations = ['한국', '미국', '영국', '프랑스', '이탈리아']
nations.reverse()
print(nations)
```

sort()
- 기본 오름차순으로 정렬
- 내림차순으로 정렬할 경우 reverse = True

```python
nations = ['한국', '미국', '영국', '프랑스', '이탈리아']
nations.sort()
print(nations)
nations.sort(reverse = True)
print(nations)
```

리스트와 랜덤

랜덤 선택
- random 라이브러리의 choic()는 리스트의 항목 한개를 무작위로 선택
 - random.choice(리스트)
- random.sample(): 설정한 개수만큼 리스트에서 항목을 중복 없이 무작위로 선택
 - random.sample(리스트, 개수)
- random.choices(리스트, k = 개수)
 - 리스트에서 항목 k개를 중복을 허용해 무작위로 선택

```python
import random

colors = ['ivory', 'lavender', 'lidhtcyan', 'lightyellow', 'palegreem']

print(random.choice(colors))

print(random.sample(colors, 3))

print(random.choices(colors, k = 3))
```

```python
import random

colors = ['ivory', 'lavender', 'lidhtcyan', 'lightyellow', 'palegreem']
random.shuffle(colors)
print(colors[:3])
```

리스트 통계 구하기
- max(): 리스트 항목 중 최대값
- min(): 리스트 항목 중 최솟값
- sum(): 리스트 항목들의 합

```python
scores = [85, 95, 70, 75, 99, 93, 88]

print('최댓값: ', max(scores))
print('최솟값: ', min(scores))
print('합: ', sum(scores))
print('평균: ', sum(scores)/len(scores))
#print(f'평균: , {sum(scores)/len(scores):, 2f}')
```
