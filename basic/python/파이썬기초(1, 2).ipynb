변수 포멧

- %변수 포멧으로 사용
- 변수 포멧을 사용하여 문자열에 변수 값을 삽입
- %s : 문자열
- %c : 문자
- %d : 정수
- %f : 실수

```python
x = 10
print("x is %d" %x)

y = "code"
print("y is %s" %y)

print("x is %d and y is %s" %(x, y))
```

format 함수 이용

- format 함수를 이용하면 변수의 타입을 명시하지 않아도 됨
- 순서 또는 변수 이름을 통해 원하는 변수를 포멧에 맞춰 넣어줘야 함

```python
x, y = 10, "code"

print("x is {0}" .format(x))
print("x is {new_x}" .format(new_x=x))
print("\n")
print("x is {0} and y is {1}" .format(x, y))
print("x is {new_x} and y is {new_y}" .format(new_x=x, new_y=y))
print("y is {1} and x is {0}" .format(x, y))
print("y is {new_y} and x is {new_x}" .format(new_x=x, new_y=y))
```

f 문자열 포멧

- 문자열 앞에 f를 붙이면, 중괄호와 변수 이름만으로 문자열에 원하는 변수를 삽입할 수 있음

```python
x, y = 10, "code"

##print("x is {0}", %x)
print("x is %d" %x)

print(f"x is {x}")
print(f"x is {y}")

print("x is {x} and y is {y}")
print(f"x is {x} and y is {y}")
```

실수형 변수를 소수점 맞춰 출력

- 4f == 소수 4째 자리까지 반올림 출력

```python
# %를 이용
x = 3.141592653
print("%.4f" %x)

# 기존 포멧 {0}에 : 을 붙여 사용
print("{0:.4f}" .format(x))

# f 문자열 포멧 이용
print(f"{x:.4f}")
```

- 남녀 비율 계산하기

```python
male_num = int(input('남성 지원자 수: '))
female_num = int(input('여성 지원자 수: '))
total = male_num + female_num
print(f'남성 비율: {male_num/total: .2f} 여성 비율: {female_num/total: .2f}')
print(f'남성 비율: {(male_num/total)*100: .2f} 여성 비율: {(female_num/total)*100: .2f}')
```

- 길이 단위 변환하기

```python
cm = int(input('센티미터 단위 길이: '))
m = cm // 100
cm = cm % 100
print(f'{m}미터 {cm}센티미터')
```

- 분 단위 시간을 시간과 분 단위로 변환하기

```python
min = int(input('분 단위 시간: '))
hour = min // 60
min = min % 60
print(f'{hour}시간 {min}분')
```

- 초 단위 시간을 시간, 분, 초 단위로 변환하기

```python
second = int(input('초 단위 시간: '))
hour = second // (60*60)
second = second % (60*60)
min = second // 60
second = second % 60
print(f'{hour}시간 {min}분 {second}초')
```

- 주문 금액 계산기1

```python
americano = int(input('아메리카노 주문 수: '))
latte = int(input('라떼 주문 수: '))
greentea = int(input('녹차 주문 수: '))
sum = americano*2000 + latte*3000 + greentea*2500
print('주문 금액:' ,sum)
```

- 주문 금액 계산기

```python
americano_price = 2000
latte_price = 3000
greentea_price = 2500

sum = 0

americano_num = int(input('아메리카노 주문 수: '))
sum += americano_num * americano_price

latte_num = int(input('라뗴 주문 수: '))
sum += latte_num * latte_price

greentea_num = int(input('녹차 주문 수: '))
sum += greentea_num * greentea_price

print('주문 금액:' ,sum)
```
