문자열 인덱싱
- 문자열의 각 문자에 접근할 때 문자 위치인 index를 사용
- 문자에 겁근하는 것을 indexing이라함

```python
s = 'abc'
for i in range(len(s)):
  print(s[i], end=' \n')
```

회문 판별
- 회문(palindrome): 문자열을 거꾸로 뒤집어도 원래 문자열과 같은 문자
- s2에 문자열 저장 후 s1 == s2 비교하여 회문 판별

```python
s1 = input('문자열: ')
s2 = ''

for i in range(len(s1)-1, -1, -1):
  s2 = s2 + s1[i]

if s1 == s2:
  print('회문')
else:
  print('회문아님')
```

대소문자 변환

- lower(): 문자열의 알파벳을 소문자로 변환한다.
- upper(): 문자열의 알파벳을 대문자로 변환한다.

```python
lower_list = []

while True:
  s = input('영어 단어 또는 exit: ')
  if s == 'exit':
    break
  lower_list.append(s.lower())

for a in lower_list:
  print(a)
```

문자열 찾기

- find(): 문자열에서 특정 문자열을 왼쪽부터 찾아서 인덱스를 반환한다. 찾는 문자열이 없으면 -1을 반환한다.
- rfind(): 문자열에서 특정 문자열을 오른쪽부터 찾아서 인덱스를 반환한다. 찾는 문자열이 없으면 -1을 반환한다.
- count(): 문자열에서 특정 문자열의 개수를 반환한다.

```python
s = input('문자열: ')
if s.find('희망') != -1:
  print('포함')
else:
  print('포함 안 됨')
```

공백 제거와 문자열 변경

- strip(): 문자열 양옆의 공백이나 특정 문자를 제거한다.
- rstrip(): 문자열 오른쪽의 공백이나 특정 문자를 제거한다.
- lstrip(): 문자열 왼쪽의 공백이나 특정 문자를 제거한다.
- replace(): 문자열에서 특정 문자열을 다른 문자로 변경한다.

```python
s1 = '  Hi! Python '
s2 = '..Hi! Python..'
print(s1)
print(s2)
```

```python
print(s1.strip())
print(s2.strip('.'))
```

```python
s = input('몸무게: ')
if s.find('kg')!=-1:
 s = s.rstrip('kg')
print(s)
```

```python
s = "###  student_name"
print(s.lstrip('#'))
```

```python
s = 'KAIST, 제주대, 홍익대'
print(s.replace('대', '대학교'))
```

문자열 분리와 결합하기

- split(): 특정 문자를 기준으로 문자열을 분리하고 리스트로 변환
- join(): 여러 문자열을 모아서 하나의 문자열로 만든다. 문자열을 구분할 구분자를 지정해야 함

```python
a = '수학, 정보'
b = '정보, 과학'
lst = a.split(', ')
lst.extend(b.split(', '))
print(lst)
```

문자열 구성 파악하기
- 사용자가 입력한 문자열이 조건에 맞는지 판별할 수 있음

- isdigit(): 문자열이 숫자로만 구성되었는지 확인한다.
- isalpha(): 문자열이 알파벳으로만 구성되었는지 확인한다.
- isalnum(): 문자열이 알파벳과 숫자로만 구성되었는지 확인한다.
- islower(): 문자열이 소문자로만 구성되었는지 확인한다.
- isupper(): 문자열이 대문자로만 구성되었는지 확인한다.
- isspace(): 문자열이 공백으로만 구성되었는지 확인한다.

아이디 입력
- 입력받은 아이디가 5자리 이상
- 알파벳과 숫자로만 이루어질 때까지 입력 받기를 반복

```python
while True:
 id = input('아이디: ')
 if len(id)<5:
   print('길이가 짧습니다.')
 elif id.isalnum()!=True:
   print('아이디는 알파벳이나 숫자로 이루어져야 합니다.')
 else:
   break
print(f'아이디는 {id}입니다.')
```

형 변환

- str(): 수를 문자열로 변환한다.
- int(): 정수 형식의 문자열이나 실수를 정수로 변환한다.
- float(): 실수 형식의 문자열이나 정수를 실수로 변환한다.

```python
s = input('키: ')

if s.find('cm')!=-1:
 height = float(s.rstrip('cm'))
else:
 height = float(s)
print(height)
```

문자 함수(ASCLL code)

- chr(): 정수에 해당하는 문자를 반환한다.
- ord(): 문자에 해당하는 정수를 반환한다.

```python
 print(chr(65))
 print(ord('A'))
```

- 소문자 2개를 입력받아 알파벳 순서를 기준으로 얼마나 떨어져 있는지 구하는 프로그램

```python
c1 = input('알파벳 소문자: ')
c2 = input('알파벳 소문자: ')

if c1 > c2:
 diff = ord(c1)-ord(c2)
else:
 diff = ord(c2)-ord(c1)
print('떨어진 정도:', diff)
```

대소문자 상호 변환
- 대문자는 소문자로, 소문자는 대문자로
 - 사용자가 입력한 문자열 대문자는 소문자로 소문자는 대문자로 변환

```python
s1 = input('문자열 : ')
s2 = ''

for ch in s1:
 if ch.islower():
   s2 += ch.upper()
 elif ch.isupper():
   s2 += ch.lower()
 else:
   s2 += ch
print(s2)
```

좋아하는 친구 조사하기
- 좋아하는 친구를 입력받아 하나의 리스트에 저장
- split 함수를 이용하여 한 명씩 분리

```python
lst = []
i = 1

while True:
 s = input(f'{i}번 학생 좋아하는 친구(,로 구분) 또는 종료: ')
 if s=='종료':
   break
 lst.extend(s.split(','))
 i += 1

 print(lst)
```

모스 부호 번역기
- 미국의 발명가 모스에 의해 발명
- 점(.) 과 선(-)의 조합으로 문자를 표기
- 자주 사용하는 문자일 수록 신호 길이가 짧음

```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
morse = ['.-','-...','-.-.','-..','.','..-.','--.',
          '....','..','.---','-.-','.-..','--','-.',
         '---','.--.','--.-','.-.','...','-','..-',
         '...-','.--','-..-','-.--','--..']

s = input('알파벳 문자열: ')

for i in range(len(s)):
 if s[i] in alphabet:
   print(morse[alphabet.index(s[i])], end=' ')
```
